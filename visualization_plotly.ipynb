```python
data = [
    go.Scatter(
        x=df['R&D Spend'],
        y=df['Administration'],
        text=df['State'],
        mode='markers',
        marker=dict(
            sizemin=10,
            size=df['R&D Spend'],
            
            showscale=True,
            line=dict(color='black', width=1.2)))
]

figure = go.Figure(
    data=data,
    layout=go.Layout(
        xaxis=dict(title='R&D Spend'),
        yaxis=dict(title='Administration')
    ))

iplot(figure)
```

```python
import plotly.figure_factory as ff

figure = ff.create_scatterplotmatrix(
    df[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit','State']],
    height=1000,
    width=1000,
    text=df['State'],
    diag='histogram',
    index='State'
    
    )
iplot(figure)
```

```python
colorscales = [
    'Greys', 'YlGnBu', 'Greens', 'YlOrRd', 'Bluered', 'RdBu', 'Reds', 'Blues',
    'Picnic', 'Rainbow', 'Portland', 'Jet', 'Hot', 'Blackbody', 'Earth',
    'Electric', 'Viridis', 'Cividis'
]
```

```python
corrs = df.corr()

figure = ff.create_annotated_heatmap(
    z=corrs.values,
    x=list(corrs.columns),
    y=list(corrs.index),
    colorscale='Earth',
    annotation_text=corrs.round(2).values,
    showscale=True, reversescale=True)

figure.layout.margin = dict(l=200, t=200)
figure.layout.height = 800
figure.layout.width = 1000

iplot(figure)
```
